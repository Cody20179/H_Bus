# Create Table

```python
MySQL_Run("""
CREATE TABLE IF NOT EXISTS `bus_route_stations` (
  `route_id` INT UNSIGNED NOT NULL COMMENT '所屬路徑ID（對應 bus_routes_total.route_id）',
  `route_name` VARCHAR(255) NOT NULL COMMENT '路徑名稱',
  `direction` ENUM('去程','回程') NOT NULL COMMENT '路程方向',
  `stop_name` VARCHAR(255) NOT NULL COMMENT '站點名稱',
  `latitude` DOUBLE NOT NULL COMMENT '緯度',
  `longitude` DOUBLE NOT NULL COMMENT '經度',
  `eta_from_start` INT UNSIGNED DEFAULT NULL COMMENT '首站到此站的預估時間（分鐘）',
  `stop_order` INT UNSIGNED NOT NULL COMMENT '第幾站（排序用）',
  `address` VARCHAR(255) DEFAULT NULL COMMENT '站點地址',
  `status` TINYINT NOT NULL DEFAULT 0 COMMENT '狀態（0=正常/啟用，可自訂）',
  `created_at` TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP COMMENT '建立時間',
  KEY `idx_route_id` (`route_id`),
  KEY `idx_route_name` (`route_name`)
)

""")
```

```python
import pymysql

# 資料庫連線設定
config = {
    "host": "10.53.1.3",
    "user": "root",
    "port": 7000,
    "password": "dh9j2(HU#s9h",
    "database": "bus_system",
    "charset": "utf8mb4"
}

# 建立資料表 SQL
create_table_sql = """
DROP TABLE IF EXISTS `reservation`;
/*!40101 SET @saved_cs_client     = @@character_set_client */;
/*!50503 SET character_set_client = utf8mb4 */;
CREATE TABLE `reservation` (
  `reservation_id` INT(11) NOT NULL AUTO_INCREMENT COMMENT '預約ID',
  `user_id` INT(11) DEFAULT NULL COMMENT '對應會員ID',
  `booking_time` TIMESTAMP NULL DEFAULT NULL COMMENT '預約時間',
  `booking_number` INT(11) UNSIGNED DEFAULT NULL COMMENT '預約人數',
  `booking_start_station_name` VARCHAR(255) COLLATE utf8mb4_unicode_ci DEFAULT NULL COMMENT '起始站名稱',
  `booking_end_station_name` VARCHAR(255) COLLATE utf8mb4_unicode_ci DEFAULT NULL COMMENT '結束站名稱',
  `payment_method` VARCHAR(50) COLLATE utf8mb4_unicode_ci DEFAULT NULL COMMENT '付款方式',
  `payment_record` TEXT COLLATE utf8mb4_unicode_ci COMMENT '付款紀錄',
  `payment_status` ENUM('pending','paid','failed','refunded') COLLATE utf8mb4_unicode_ci DEFAULT 'pending' COMMENT '付款狀態',
  `review_status` ENUM('pending','approved','rejected') COLLATE utf8mb4_unicode_ci DEFAULT 'pending' COMMENT '審核狀態 (人工管理)',
  `dispatch_status` ENUM('not_assigned','assigned') COLLATE utf8mb4_unicode_ci DEFAULT 'not_assigned' COMMENT '派車狀態',
  `reservation_status` TINYINT(1) GENERATED ALWAYS AS (
      IF(payment_status = 'paid' AND review_status = 'approved' AND dispatch_status = 'assigned', 1, 0)
  ) STORED COMMENT '預約狀態 (付款完成且審核通過才成立)',
  `created_at` TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP COMMENT '建立時間',
  `updated_at` TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP ON UPDATE CURRENT_TIMESTAMP COMMENT '更新時間',
  PRIMARY KEY (`reservation_id`),
  KEY `idx_user_id` (`user_id`),
  KEY `idx_booking_time` (`booking_time`),
  CONSTRAINT `fk_reservation_user` FOREIGN KEY (`user_id`) 
    REFERENCES `users` (`user_id`) 
    ON DELETE SET NULL 
    ON UPDATE CASCADE
) ENGINE=InnoDB 
  DEFAULT CHARSET=utf8mb4 
  COLLATE=utf8mb4_unicode_ci;
  """
# 執行
try:
    conn = pymysql.connect(**config)
    cursor = conn.cursor()
    for statement in create_table_sql.strip().split(";"):
        if statement.strip():
            cursor.execute(statement)
    conn.commit()
    print("✅ reservation 表建立完成")
except Exception as e:
    print("❌ 錯誤：", e)
finally:
    cursor.close()
    conn.close()
```
